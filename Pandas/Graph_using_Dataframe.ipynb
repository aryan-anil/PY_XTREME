```python
import pandas as pd
y = {'f(x)':[3,7,4,9,2],'g(x)':[2,5,3,7,5]} #Y basically contains two set of values representing two functions
x = [1,3,4,5,7]   # X is the set of independent values 
graph = pd.DataFrame(y,x)   #load data into a DataFrame object
print("DataFrame Table")  
print(graph)  
# Plotting the values into the graph
graph.plot(kind='line' , grid=True , title="my graph" , xlabel="X-axis ->" , ylabel="Y-axis ->")
```
