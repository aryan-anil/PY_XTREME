```python
from matplotlib import pyplot as plt
from matplotlib import style
 
style.use('ggplot')
x = [5,8,10]
y = [7,10,6]
x2 = [6,9,11]
y2 = [6,10,7]
plt.plot(x,y,'r',label='product 1', linewidth=5)
plt.plot(x2,y2,'b',label='product 2',linewidth=5)
plt.title('Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.legend()
plt.grid(True,color='k')
plt.show()
```
